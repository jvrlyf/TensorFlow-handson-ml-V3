```python
import tensorflow as tf
from tensorflow import keras

```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1: Fetch the California housing dataset from sklearn's library
housing = fetch_california_housing()
```

```python
# Step 2: Split the data into a training set and a test set (default 75/25 split)
x_train_full, x_test, y_train_full, y_test = train_test_split(housing.data, housing.target)

# Step 3: Split the training set again to create a validation set for monitoring performance during training
x_train, x_valid, y_train, y_valid = train_test_split(x_train_full, y_train_full)
```

```python
# Step 4: Initialize the StandardScaler to normalize our features (mean=0, variance=1)
scalar = StandardScaler()
```

```python
# Step 5: Fit the scaler on training data and transform all subsets
# Note: Use fit_transform only on training data, and transform on validation/test
x_train_scaled = scalar.fit_transform(x_train)
x_valid_scaled = scalar.transform(x_valid)
x_test_scaled = scalar.transform(x_test)
```

<h2>
<font color="#2E7D32">
Key Concepts: Regression vs. Classification MLPs
</font>
</h2>

<p>
When building a
<b><font color="#D32F2F">Regression MLP</font></b>
using the
<b><font color="#1565C0">Sequential API</font></b>,
there are several important differences compared to
<b><font color="#F57C00">Classification MLPs</font></b>.
Understanding these differences helps in designing an effective neural network for predicting
<b><font color="#388E3C">continuous numerical values</font></b>.
</p>

<hr>

<h3>
<font color="#1976D2">
Key Differences
</font>
</h3>

<ul>

<li>
<b><font color="#F57C00">Output Layer:</font></b><br>
Regression models predict a
<b>single continuous numerical value</b>.
Therefore, the output layer contains
<b>one neuron</b>
and generally uses
<b>no activation function</b>
(<code>activation=None</code> or linear activation),
allowing predictions across the entire real number line.
</li>

<br>

<li>
<b><font color="#388E3C">Loss Function:</font></b><br>
Instead of classification losses such as
<code>categorical_crossentropy</code>,
regression models commonly use
<b>Mean Squared Error (MSE)</b>.
MSE penalizes larger prediction errors more heavily, making it well-suited for numerical prediction tasks.
</li>

<br>

<li>
<b><font color="#1565C0">Architecture Design:</font></b><br>
Real-world regression datasets often contain
<b>noise</b>.
To reduce the risk of
<b>overfitting</b>,
it is generally recommended to begin with a
<b>simple network architecture</b>,
such as a single hidden layer with a moderate number of neurons, before increasing model complexity.
</li>

</ul>

<hr>

<p align="center">
<b><font color="#C62828">
Takeaway:
</font></b>
Regression MLPs differ from classification networks primarily in their
<b>output layer</b>,
<b>loss function</b>,
and
<b>model complexity</b>,
all of which are tailored for predicting continuous numerical values.
</p>

```python
model = keras.models.Sequential([
    keras.layers.Dense(30 , activation = "relu", input_shape = x_train.shape[1:]),
    keras.layers.Dense(1)
])
```

```python
model.compile(loss = "mean_squared_error", optimizer="sgd")
```

```python
history = model.fit(x_train, y_train , epochs=20, validation_data=(x_valid, y_valid))
```

```python
mse_test = model.evaluate(x_test, y_test)
```

```python
# pretend these are new instances
x_new = x_test[:3]
```

```python
y_pred = model.predict(x_new)
```

<h2>
<font color="#2E7D32">
Building Complex Models Using the Functional API
</font>
</h2>

<p>
The
<b><font color="#1565C0">Sequential API</font></b>
is ideal for building simple neural networks where layers are connected in a
<b><font color="#388E3C">linear sequence</font></b>.
However, many real-world deep learning problems require more flexible architectures.
For these scenarios,
<b><font color="#D32F2F">Keras Functional API</font></b>
provides a powerful way to design complex neural networks.
</p>

<hr>

<h3>
<font color="#1976D2">
Why Use the Functional API?
</font>
</h3>

<p>
Unlike the Sequential API, where each layer has exactly
<b>one input</b>
and
<b>one output</b>,
the Functional API allows layers to be connected in
<b>multiple directions</b>,
making it possible to build advanced neural network architectures.
</p>

<ul>

<li>
<b><font color="#F57C00">Multi-Input Models:</font></b><br>
Combine different types of input data, such as numerical features, images, or text, within a single model.
</li>

<br>

<li>
<b><font color="#388E3C">Multi-Output Models:</font></b><br>
Predict multiple target variables simultaneously using separate output layers.
</li>

<br>

<li>
<b><font color="#1565C0">Non-Linear Topologies:</font></b><br>
Create networks with branching, merging, skip connections, or shared layers that cannot be represented using the Sequential API.
</li>

</ul>

<hr>

<h3>
<font color="#1976D2">
Wide &amp; Deep Learning Architecture
</font>
</h3>

<p>
A well-known example of a
<b><font color="#D32F2F">non-sequential architecture</font></b>
is the
<b><font color="#1565C0">Wide &amp; Deep Neural Network</font></b>,
introduced by
<b>Heng-Tze Cheng et al.</b>
in 2016.
This architecture combines the strengths of both
<b>memorization</b>
and
<b>generalization</b>
within a single model.
</p>

<h4>
<font color="#388E3C">
How It Works
</font>
</h4>

<ul>

<li>
<b><font color="#F57C00">Wide Path (Short Path):</font></b><br>
Some or all of the input features are connected
<b>directly to the output layer</b>.
This enables the model to learn simple relationships and memorize important patterns without passing through multiple hidden layers.
</li>

<br>

<li>
<b><font color="#1565C0">Deep Path:</font></b><br>
The same input features are also passed through several
<b>hidden layers</b>,
allowing the network to learn complex, high-level feature representations and nonlinear relationships.
</li>

</ul>

<hr>

<h3>
<font color="#1976D2">
Advantages of Wide &amp; Deep Models
</font>
</h3>

<ul>

<li>
<b><font color="#388E3C">Memorization:</font></b>
The wide path captures simple and previously observed relationships.
</li>

<br>

<li>
<b><font color="#388E3C">Generalization:</font></b>
The deep path discovers complex patterns and interactions between features.
</li>

<br>

<li>
<b><font color="#388E3C">Improved Performance:</font></b>
Combining both paths often leads to better predictive performance than using either approach alone.
</li>

</ul>

<hr>

<p align="center">
<b><font color="#C62828">
Takeaway:
</font></b>
The
<b>Functional API</b>
enables the creation of advanced neural network architectures, including
<b>multi-input</b>,
<b>multi-output</b>,
and
<b>Wide &amp; Deep</b>
models, providing greater flexibility than the Sequential API for solving real-world machine learning problems.
</p>

<h1>
<font color="#1565C0">
Implementation Details: Keras Functional API
</font>
</h1>

<hr>

<p>
At a high level, the first five lines create all the layers we need to build the
<b><font color="#1565C0">model</font></b>, the next six lines use these layers just like functions to go from the
<b><font color="#388E3C">input</font></b> to the <b><font color="#D32F2F">output</font></b>, and the last line creates a
<code>Keras Model</code> object by pointing to the input and the output.
</p>

<u>The Functional API allows for complex model construction by treating layers as functions.</u>

```python
normalization_layer = tf.keras.layers.Normalization()
hidden_layer1 = tf.keras.layers.Dense(30, activation = 'relu')
hidden_layer2 = tf.keras.layers.Dense(30, activation = 'relu')

concat_layer = tf.keras.layers.Concatenate()

output_layer = tf.keras.layers.Dense(1)

# Fix: use [1:] to get the number of features (8) instead of sample count
input_ = tf.keras.layers.Input(shape = x_train.shape[1:])

normalized = normalization_layer(input_)
hidden1 = hidden_layer1(normalized)
hidden2 = hidden_layer2(hidden1)

concat = concat_layer([normalized, hidden2])

output = output_layer(concat)
```

```python
model = tf.keras.Model(inputs=[input_], outputs=[output])
```

```python
model.summary()
```

<h2>
<font color="#2E7D32">
Understanding the Keras Functional API
</font>
</h2>

<p>
At a high level, the
<b><font color="#1565C0">first five lines</font></b>
create all the
<b><font color="#1565C0">layers</font></b>
required to build the model.
The
<b><font color="#1565C0">next six lines</font></b>
connect these layers together like functions, allowing information to flow from the
<b><font color="#388E3C">input</font></b>
to the
<b><font color="#388E3C">output</font></b>.
Finally, the
<b><font color="#D32F2F">last line</font></b>
creates a
<b><font color="#D32F2F">Keras Model</font></b>
by specifying its inputs and outputs.
</p>

<hr>

<h3>
<font color="#2E7D32">
Step 1: Creating the Layers
</font>
</h3>

<p>
The model starts by creating
<b><font color="#1565C0">five layers</font></b>:
</p>

<ul>
<li>A <b><font color="#388E3C">Normalization</font></b> layer to standardize the input features.</li>

<li>Two <b><font color="#388E3C">Dense layers</font></b>, each containing <b>30 neurons</b> and using the <b>ReLU activation function</b>.</li>

<li>A <b><font color="#388E3C">Concatenate layer</font></b> that combines outputs from different parts of the network.</li>

<li>A final <b><font color="#388E3C">Dense layer</font></b> with <b>one neuron</b> that produces the model's output without using an activation function.</li>
</ul>

<hr>

<h3>
<font color="#2E7D32">
Step 2: Creating the Input Object
</font>
</h3>

<p>
Next, an
<b><font color="#1565C0">Input object</font></b>
is created. The variable name
<b>input_</b>
is used instead of
<b>input</b>
to avoid conflicting with Python's built-in
<b>input()</b>
function.
</p>

<p>
The Input object defines:
</p>

<ul>
<li>The <b><font color="#388E3C">shape</font></b> of the input data.</li>

<li>The <b><font color="#388E3C">data type (dtype)</font></b>, which is <b>32-bit floating-point</b> by default.</li>

<li>A model can also contain <b><font color="#D32F2F">multiple inputs</font></b> when needed.</li>
</ul>

<hr>

<h3>
<font color="#2E7D32">
Step 3: Connecting the Layers
</font>
</h3>

<p>
The
<b><font color="#1565C0">Normalization layer</font></b>
is called like a function by passing the
<b>Input object</b>
to it. This is why it is called the
<b><font color="#D32F2F">Functional API</font></b>.
</p>

<p>
At this stage,
<b><font color="#D32F2F">no actual data is processed</font></b>.
The Input object is only a
<b>symbolic input</b>
that describes the expected data. Likewise, the output stored in
<b>normalized</b>
is also symbolic and is used only to construct the model.
</p>

<p>
The symbolic outputs are connected as follows:
</p>

<ul>
<li><b>normalized → hidden_layer1 → hidden1</b></li>

<li><b>hidden1 → hidden_layer2 → hidden2</b></li>
</ul>

<p>
Each layer receives the symbolic output from the previous layer and produces another symbolic output, gradually building the model architecture.
</p>

<hr>

<h3>
<font color="#2E7D32">
Step 4: Concatenating the Layers
</font>
</h3>

<p>
So far, all layers are connected in a
<b><font color="#1565C0">sequential manner</font></b>.
The
<b><font color="#388E3C">Concatenate layer</font></b>
then combines:
</p>

<ul>
<li>The original <b>input</b>.</li>

<li>The output of the <b>second hidden layer</b>.</li>
</ul>

<p>
Again,
<b><font color="#D32F2F">no actual data is concatenated</font></b>.
The operation remains symbolic and simply defines how the layers should be connected.
</p>

<hr>

<h3>
<font color="#2E7D32">
Step 5: Producing the Final Output
</font>
</h3>

<p>
The concatenated output is passed to the
<b><font color="#388E3C">final Dense layer</font></b>,
which generates the model's
<b><font color="#1565C0">final prediction</font></b>.
</p>

<hr>

<h3>
<font color="#2E7D32">
Step 6: Creating the Keras Model
</font>
</h3>

<p>
Finally, a
<b><font color="#D32F2F">Keras Model</font></b>
is created by specifying:
</p>

<ul>
<li>The model's <b><font color="#388E3C">input(s)</font></b>.</li>

<li>The model's <b><font color="#388E3C">output(s)</font></b>.</li>
</ul>

<p>
Once the model is created, the architecture is complete and is ready for
<b><font color="#1565C0">compilation</font></b>,
<b><font color="#1565C0">training</font></b>, and
<b><font color="#1565C0">making predictions</font></b>.
</p>

<h2>
<font color="#2E7D32">
Using the Functional API Model
</font>
</h2>

<p>
Once the
<b><font color="#1565C0">Keras model</font></b>
is built, the remaining steps are exactly the same as before.
You simply
<b><font color="#388E3C">compile</font></b>
the model,
<b><font color="#388E3C">adapt</font></b>
the
<b>Normalization</b>
layer,
<b><font color="#388E3C">train (fit)</font></b>
the model,
<b><font color="#388E3C">evaluate</font></b>
its performance, and finally
<b><font color="#388E3C">make predictions</font></b>
using new data.
</p>

<hr>

<h2>
<font color="#2E7D32">
Using Multiple Inputs
</font>
</h2>

<p>
Sometimes, instead of sending
<b><font color="#1565C0">all input features</font></b>
through the same path, we may want to send
<b><font color="#388E3C">different groups of features</font></b>
through different paths in the network.
This can be done by using
<b><font color="#D32F2F">multiple inputs</font></b>.
</p>

<p>
For example, suppose a dataset has
<b>8 features</b>
(numbered from
<b>0</b>
to
<b>7</b>):
</p>

<ul>
<li>
The
<b><font color="#388E3C">wide path</font></b>
receives
<b>Features 0 to 4</b>.
</li>

<li>
The
<b><font color="#388E3C">deep path</font></b>
receives
<b>Features 2 to 7</b>.
</li>
</ul>

<p>
Notice that
<b><font color="#1565C0">Features 2, 3, and 4</font></b>
are used in both paths.
This allows the model to learn
<b><font color="#D32F2F">different patterns</font></b>
from the same data, making it more flexible and powerful.
</p>

```python
input_wide = tf.keras.layers.Input(shape = [5]) # 0-4 wide features
input_deep = tf.keras.layers.Input(shape = [6]) # 2-7 deep features
```

```python
norm_layer_wide = tf.keras.layers.Normalization()
norm_layer_deep = tf.keras.layers.Normalization()
```

```python
norm_wide = norm_layer_wide(input_wide)
norm_deep = norm_layer_deep(input_deep)
```

```python
hidden1 = tf.keras.layers.Dense(30, activation = 'relu')(norm_deep)
hidden2 = tf.keras.layers.Dense(30 , activation = 'relu')(hidden1)
concat = tf.keras.layers.concatenate([input_wide, hidden2])
output = tf.keras.layers.Dense(1)(concat)

model = tf.keras.Model(inputs=[input_wide, input_deep], outputs=[output])
```

<h2>
<font color="#2E7D32">
Important Points About This Example
</font>
</h2>

<ul>

<li>
Each
<b><font color="#1565C0">Dense layer</font></b>
is <b>created and called on the same line</b>.
This is a common practice because it makes the code
<b><font color="#388E3C">shorter and easier to read</font></b>
without reducing clarity.
</li>

<br>

<li>
The
<b><font color="#1565C0">Normalization layer</font></b>
is created separately because we need to call its
<b>adapt()</b>
method before training the model.
Keeping a reference to this layer is therefore necessary.
</li>

<br>

<li>
Instead of creating a
<b>Concatenate</b>
layer separately, we use
<b><font color="#1565C0">tf.keras.layers.concatenate()</font></b>,
which automatically
<b><font color="#388E3C">creates the Concatenate layer and combines the given inputs</font></b>.
</li>

<br>

<li>
Since the model has
<b><font color="#D32F2F">two input layers</font></b>,
we specify them while creating the model using:
<b>inputs = [input_wide, input_deep]</b>.
</li>

<br>

<li>
The model is
<b><font color="#388E3C">compiled</font></b>
in the usual way.
However, while calling
<b>fit()</b>,
we must provide
<b><font color="#1565C0">two input datasets</font></b>
instead of one:
<b>(X_train_wide, X_train_deep)</b>.
Each dataset is sent to its corresponding input layer.
</li>

<br>

<li>
The same idea applies when using
<b><font color="#388E3C">evaluate()</font></b>
or
<b><font color="#388E3C">predict()</font></b>.
Instead of passing a single dataset, we pass
<b><font color="#1565C0">two input datasets</font></b>,
such as
<b>(X_valid_wide, X_valid_deep)</b>,
<b>(X_test_wide, X_test_deep)</b>,
or
<b>(X_new_wide, X_new_deep)</b>.
</li>

</ul>
